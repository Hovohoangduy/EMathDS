```python
# Ex 1-17. A derivate calculator in Python
def derivate_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_func(x):
    return x**2

slop_at_2 = derivate_x(my_func, 2, .00001)
print(slop_at_2)
```

```python
# Ex 1-25. Calculating the derivative dz/dx with and without the chain rule, but still getting the same answer
from sympy import *

x, y = symbols('x y')
_y = x**2 + 1
dy_dx  = diff(_y)
z = y**3 - 2
dz_dy = diff(z)
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

print(dz_dx_chain)
print(dz_dx_no_chain)
```

```python
# Ex 1-26. An integral approximation in Python
def approxima_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
    return total_sum * delta_x

def my_func(x):
    return x**2 + 1

area = approxima_integral(a = 0, b=1, n=5, f=my_func)
print(area)
```

```python
# Ex 1-27. Another integral appoximation in Python
area = approxima_integral(a=0, b=1, n=1000, f=my_func)
print(area)
```

```python
# Ex 1-28. Yet another integral approximation in Python
area = approxima_integral(a=0, b=1, n=1_000_000, f=my_func)
print(area)
```

```python
# Ex 1-30. Using limits to calculate integrals
from sympy import * 

x, i, n = symbols('x i n')
f = x**2 + 1
lower, upper = 0, 1
# Calculate with and each rectangle height at index "i"
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)
# Interate all "n" rectangle and sum their areas
n_retangles = Sum(delta_x * fx_i, (i, 1, n)).doit()
area = limit(n_retangles, n, oo)
print(area)
```
